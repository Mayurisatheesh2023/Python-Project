```python

subscription_counts = netflix_df['Subscription Type'].value_counts()
```

```python

plt.figure(figsize=(8, 8))
plt.pie(subscription_counts, labels=subscription_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Netflix Subscription Types')
plt.show()
```

```python
country_counts = netflix_df['Country'].value_counts()

```

```python
plt.pie(country_counts, labels=country_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Country Counts')
plt.show()
```

```python
device_counts = netflix_df["Device"].value_counts()
```

```python
device_counts.plot(kind='bar', color=['red','green','orange','blue'])
plt.xlabel('Device')
plt.ylabel('Count')
plt.title('Count of Each Device Type')
plt.show()
```

```python
age_counts = netflix_df['Age'].value_counts()
```

```python
age_counts
```

```python
plt.figure(figsize=(7, 5))
plt.hist(netflix_df["Age"], bins=20, color='green', edgecolor='black')  # You can adjust the number of bins
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Ages')
plt.show()
```

```python
gender_counts = netflix_df['Gender'].value_counts()
gender_counts
```

```python
plt.figure(figsize=(7, 7))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90, colors=['blue', 'pink'])
plt.title('Distribution of Genders')
plt.show()
```

```python
subscription_revenue = netflix_df.groupby('Subscription Type')['Monthly Revenue'].sum()
plt.figure(figsize=(10, 6))
subscription_revenue.plot(kind='bar', color='skyblue', edgecolor='black')
plt.xlabel('Subscription Type')
plt.ylabel('Revenue ($)')
plt.title('Revenue Analysis by Subscription Type')
```
